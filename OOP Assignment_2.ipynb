```python
"""Q1. Which function is used to open a file? What are the different modes of opening a file? Explain each mode
of file opening.
In Python, the built-in function used to open a file is open().
Syntax: 'open(file_path, mode)'
Here, file_path represents the path or name of the file that you want to open, and mode specifies the mode in which you want to open the 
file.The mode parameter is optional and if you don't specify it, open() will assume the default mode of 'r' (read-only mode).
There are several modes in which a file can be opened in Python. Here are the most common ones:
1.'r' (Read mode): This mode is used to read the contents of a file. The file is opened in read-only mode and if the file does not exist, 
a FileNotFoundError is raised.

2.'w' (Write mode): This mode is used to write data to a file. If the file already exists, it will be overwritten. If it does not exist, a 
new file will be created.

3.'a' (Append mode): This mode is used to append data to an existing file. If the file does not exist, a new file will be created.

4.'x' (Exclusive mode): This mode is used to create a new file, but it will raise a FileExistsError if the file already exists.

5.'b' (Binary mode): This mode is used to open a file in binary mode. This mode should be used when working with non-text files, such 
as images or audio files.

6.'t' (Text mode): This mode is used to open a file in text mode, which is the default mode. It is used when working with text files.

You can use these modes in various combinations, such as 'rb' for reading a binary file or 'w+' for both reading and writing a text file.
```

```python
Q2. Why close() function is used? Why is it important to close a file?
Ans.:In Python, the close() function is used to close a file after it has been opened using the open() function.
Syntax: file_object.close().It is important to close a file because when a file is opened, it occupies some resources such as memory 
and CPU time, and keeping it open unnecessarily can lead to resource wastage. Closing a file releases these resources and makes them 
available for other processes.In addition to freeing up resources, closing a file ensures that any changes made to the file are properly 
saved. For example, if you open a file in write mode and write some data to it, the data will not be saved until you close the file. 
If you don't close the file, the changes made to it may be lost.
```

```python
"""Q3. Write a python program to create a text file. Write ‘I want to become a Data Scientist’ in that file. Then
close the file. Open this file and read the content of the file.
Ans.:"""

# Open the file in write mode
file = open("my_file.txt", "w")

# Write "I want to become a Data Scientist" to the file
file.write("I want to become a Data Scientist")

# Close the file
file.close()

# Open the file in read mode
file = open("my_file.txt", "r")

# Read the contents of the file
content = file.read()

# Print the contents of the file
print(content)

# Close the file
file.close()
```

```python
"""Q4. Explain the following with python code: read(), readline() and readlines().
Ans.:In Python, there are several functions available for reading the contents of a file. Here's an explanation of three commonly 
used functions: read(), readline(), and readlines().
"read()"
The read() function is used to read the entire contents of a file at once.
Syntax: file_object.read(size)
Here, size is an optional parameter that specifies the number of bytes to be read from the file. If you don't specify size, the entire 
contents of the file will be read."""

# Open the file in read mode
file = open("my_file.txt", "r")

# Read the entire contents of the file
content = file.read()

# Close the file
file.close()

# Print the contents of the file
print(content)

"""
'readline()'
The readline() function is used to read a single line from a file at a time.
Syntax: file_object.readline(size)
Here, size is an optional parameter that specifies the maximum number of characters to be read from the line. If you don't specify 
size, the entire line will be read."""

# Open the file in read mode
file = open("my_file.txt", "r")

# Read the first line of the file
line = file.readline()

# Close the file
file.close()

# Print the first line of the file
print(line)


"""
'readlines()'
The readlines() function is used to read all the lines of a file and return them as a list.
Syntax: 'file_object.readlines(sizehint)'
Here, sizehint is an optional parameter that specifies the maximum number of bytes to be read from the file. If you don't specify sizehint,
the entire contents of the file will be read."""

# Open the file in read mode
file = open("my_file.txt", "r")

# Read all the lines of the file
lines = file.readlines()

# Close the file
file.close()

# Print all the lines of the file
for line in lines:
    print(line)
```

```python
"""Q6. Explain the write() and writelines() functions. Give a suitable example.
Ans.:The write() and writelines() functions are used to write data to a file in Python.
1.'write()'
The write() function is used to write a single string to a file. If the file is opened in w or a mode, the function overwrites or appends 
the existing content respectively. file_object.write(string)
Example.: """

# Open the file in write mode
file_object = open("my_file.txt", "w")

# Write a string to the file
file_object.write("Hello, World!")

# Close the file
file_object.close()

"""2.'writelines()'
The writelines() function, on the other hand, is used to write a list of strings to a file. Each string in the list is written to the 
file as a separate line.file_object.writelines(list_of_strings)"""
# Open the file in write mode
file_object = open("my_file.txt", "w")

# Write a list of strings to the file
list_of_strings = ["apple\n", "banana\n", "cherry\n"]
file_object.writelines(list_of_strings)

# Close the file
file_object.close()
```
