```python
Q1: What is Estimation Statistics? Explain point estimate and interval estimate.
Ans.: Estimation statistics is a branch of statistics that involves estimating unknown population parameters based on sample data. When we don't have
access to the entire population, we rely on statistical inference to make educated guesses about population characteristics.

Point Estimate:
A point estimate is a single value that is used to estimate an unknown population parameter. It is based on sample data and provides an approximation 
of the true value of the parameter. For example, if we want to estimate the mean height of a population, we can calculate the mean height of a sample 
and use it as a point estimate for the population mean.

Interval Estimate:
An interval estimate is a range of values within which the population parameter is believed to lie. It takes into account the variability and 
uncertainty associated with the estimation process. The interval estimate is constructed based on the sample data and a chosen level of confidence. 
The level of confidence determines the probability that the interval contains the true parameter value.

In interval estimation, we provide a lower and an upper bound for the parameter. For example, a 95% confidence interval for the population mean 
height might be [165 cm, 175 cm]. This means that we are 95% confident that the true population mean height falls between 165 cm and 175 cm.

The width of the interval estimate is influenced by factors such as the sample size, variability of the data, and the chosen level of confidence. 
A higher level of confidence will result in a wider interval, as we want to be more certain that we capture the true parameter value.
```

```python
"""Q2. Write a Python function to estimate the population mean using a sample mean and standard
deviation.
Ans.: """
    
    import math

def estimate_population_mean(sample_mean, sample_std_dev, sample_size):
    standard_error = sample_std_dev / math.sqrt(sample_size)
    margin_of_error = 1.96 * standard_error  # Assuming a 95% confidence level
    
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return (lower_bound, upper_bound)
```

```python
Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.
Ans.: Hypothesis testing is a statistical procedure used to make inferences or decisions about population parameters based on sample data. It involves
formulating two competing hypotheses, the null hypothesis (H0) and the alternative hypothesis (H1), and collecting evidence from the data to support 
or reject the null hypothesis.

The null hypothesis typically represents the status quo or the absence of an effect, while the alternative hypothesis represents the claim or effect 
we want to investigate. Hypothesis testing helps us evaluate whether the evidence from the sample supports the alternative hypothesis or if it is
more in line with the null hypothesis.

Importance of Hypothesis Testing:

Evidence-Based Decision Making: Hypothesis testing provides a systematic framework for making decisions based on empirical evidence rather than 
subjective opinions or intuition. It allows us to objectively evaluate claims and draw conclusions from data.

Scientific Research: Hypothesis testing is a fundamental tool in scientific research. It helps researchers investigate research questions, validate 
theories, and contribute to the body of knowledge by providing empirical evidence.

Statistical Inference: Hypothesis testing is a key component of statistical inference. By testing hypotheses about population parameters, we can make 
inferences about the larger population based on the sample data. This allows us to generalize our findings and draw conclusions beyond the observed 
sample.

Quality Control and Process Improvement: Hypothesis testing is widely used in industries for quality control and process improvement purposes. By 
setting up hypotheses about process parameters, organizations can evaluate whether their processes are operating within acceptable limits and take 
appropriate actions for improvement.

Decision Making in Business and Policy: Hypothesis testing provides a rigorous approach for decision making in business and policy settings. It 
allows organizations to test the effectiveness of new strategies, marketing campaigns, or policy interventions before implementing them on a larger 
scale.
```

```python
"""Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
given a sample from each population.
Ans.: """
    
    import scipy.stats as stats

def conduct_hypothesis_test(sample1, sample2, alpha):
    # Perform independent samples t-test
    t_statistic, p_value = stats.ttest_ind(sample1, sample2)

    # Compare p-value with the significance level (alpha) to make a decision
    if p_value < alpha:
        print("Reject the null hypothesis. There is a significant difference between the population means.")
    else:
        print("Fail to reject the null hypothesis. There is not enough evidence to conclude a significant difference between the population means.")

    # Return the t-statistic and p-value
    return t_statistic, p_value
```

```python
Q6: What is a null and alternative hypothesis? Give some examples.
Ans.: In statistical hypothesis testing, the null hypothesis (H0) and the alternative hypothesis (H1 or Ha) are two competing statements about a 
population parameter or a relationship between variables. These hypotheses help guide the statistical analysis and inference process.

The null hypothesis represents the status quo or the assumption of no effect, no difference, or no relationship. It is often denoted as a statement 
of equality or no change. The objective of hypothesis testing is to assess whether the evidence from the data supports or contradicts the null 
hypothesis.

The alternative hypothesis, on the other hand, represents an alternative claim or effect that the researcher or analyst wants to investigate. It 
suggests that there is an effect, difference, or relationship in the population that is different from what is assumed by the null hypothesis.

Here are some examples of null and alternative hypotheses:

Example: Testing the effectiveness of a new drug
Null Hypothesis (H0): The new drug has no effect or is not different from the existing treatment.
Alternative Hypothesis (H1): The new drug is effective or shows a significant improvement over the existing treatment.

Example: Evaluating the impact of a training program
Null Hypothesis (H0): There is no difference in performance between individuals who underwent the training program and those who did not.
Alternative Hypothesis (H1): Individuals who underwent the training program perform better than those who did not.

Example: Investigating the relationship between smoking and lung cancer
Null Hypothesis (H0): There is no association between smoking and the incidence of lung cancer.
Alternative Hypothesis (H1): Smoking is associated with an increased incidence of lung cancer.

Example: Comparing the average incomes of two groups
Null Hypothesis (H0): The average income of Group A is equal to the average income of Group B.
Alternative Hypothesis (H1): The average income of Group A is different from the average income of Group B.
```

```python
Q7: Write down the steps involved in hypothesis testing.
Ans.: The steps involved in hypothesis testing can be summarized as follows:

Formulate the Research Hypotheses: Clearly state the null hypothesis (H0) and the alternative hypothesis (H1 or Ha) based on the research question 
or problem.

Set the Significance Level: Determine the desired level of significance (alpha) for the hypothesis test. It represents the probability of rejecting 
the null hypothesis when it is true. Commonly used significance levels are 0.05 (5%) or 0.01 (1%).

Select the Test Statistic: Choose an appropriate test statistic based on the type of data and the specific hypothesis being tested. Examples include 
the t-statistic, z-statistic, chi-square statistic, or F-statistic.

Determine the Distribution: Identify the appropriate probability distribution for the test statistic under the null hypothesis. This is necessary to 
calculate the p-value or to compare the test statistic with critical values.

Collect and Analyze the Data: Collect a sample of data and perform the necessary calculations or statistical analysis based on the chosen test 
statistic.

Calculate the Test Statistic: Compute the test statistic using the collected sample data and any relevant parameters.

Determine the Rejection Region/Critical Value: Determine the critical value(s) or the rejection region for the test statistic based on the chosen 
significance level and the distribution. This defines the boundary beyond which the null hypothesis will be rejected.

Calculate the p-value: If using a p-value approach, calculate the probability of obtaining a test statistic as extreme or more extreme than the 
observed test statistic, assuming the null hypothesis is true.

Make a Decision: Compare the test statistic with the critical value(s) or the p-value with the significance level (alpha). If the test statistic 
falls within the rejection region or the p-value is smaller than alpha, reject the null hypothesis. Otherwise, if the test statistic is outside the 
rejection region or the p-value is greater than or equal to alpha, fail to reject the null hypothesis.

Draw Conclusions: Based on the decision, interpret the results in the context of the research question and draw appropriate conclusions. If the null 
hypothesis is rejected, provide evidence supporting the alternative hypothesis. If the null hypothesis is not rejected, acknowledge the lack of 
evidence for the alternative hypothesis.
```

```python
Q8. Define p-value and explain its significance in hypothesis testing.
Ans.: In hypothesis testing, the p-value is a measure of the strength of the evidence against the null hypothesis (H0). It quantifies the probability
of obtaining a test statistic as extreme as, or more extreme than, the observed value, assuming that the null hypothesis is true.

The significance of the p-value lies in its role in decision making. It helps determine whether to reject or fail to reject the null hypothesis based
on the observed data. The general interpretation of the p-value is as follows:

If the p-value is less than or equal to the significance level (alpha) chosen for the test (e.g., 0.05), it is considered statistically significant.
In this case, we have sufficient evidence to reject the null hypothesis and support the alternative hypothesis. It suggests that the observed data 
is unlikely to have occurred by chance under the assumption of the null hypothesis.

If the p-value is greater than the significance level (alpha), it is considered statistically non-significant. In this case, we fail to reject the 
null hypothesis, indicating that the observed data does not provide strong enough evidence to support the alternative hypothesis. However, it does 
not necessarily imply that the null hypothesis is true; rather, it suggests that the evidence is inconclusive or insufficient to make a significant
conclusion.

Key points to consider regarding the p-value:

The p-value is not a measure of the size or magnitude of an effect; it solely provides information about the strength of evidence against the null 
hypothesis.A smaller p-value does not indicate the practical significance or importance of a result; it only suggests that the observed data is
unlikely to have occurred by chance.The p-value does not provide information about the alternative hypothesis being true or the probability of the alternative hypothesis; it focuses on 
the null hypothesis.
```

```python
"""Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom
parameter set to 10.
Ans.:""" 
    
    import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Set the degrees of freedom
df = 10

# Generate x values for the t-distribution plot
x = np.linspace(t.ppf(0.001, df), t.ppf(0.999, df), 100)

# Calculate the probability density function (pdf) for the given degrees of freedom
pdf = t.pdf(x, df)

# Plot the t-distribution
plt.plot(x, pdf, 'b-', label='t-distribution (df=10)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title("Student's t-Distribution")
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
random samples of equal size and a null hypothesis that the population means are equal.
Ans.:"""
    
     import scipy.stats as stats

def two_sample_t_test(sample1, sample2):
    # Perform two-sample t-test
    t_statistic, p_value = stats.ttest_ind(sample1, sample2)

    # Print the test statistic and p-value
    print("Test Statistic: ", t_statistic)
    print("p-value: ", p_value)

    # Compare p-value with significance level (e.g., alpha = 0.05) to make a decision
    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis. There is a significant difference between the population means.")
    else:
        print("Fail to reject the null hypothesis. There is not enough evidence to conclude a significant difference between the population means.")


# Example usage
sample1 = [1, 2, 3, 4, 5]  # First sample
sample2 = [2, 4, 6, 8, 10]  # Second sample

two_sample_t_test(sample1, sample2)
```

```python
Q11: What is Student’s t distribution? When to use the t-Distribution.
Ans.: Student's t-distribution, often referred to as the t-distribution, is a probability distribution that is used in hypothesis testing and 
confidence interval estimation when the sample size is small or when the population standard deviation is unknown.

The t-distribution is similar to the standard normal distribution (Z-distribution) but has thicker tails. The shape of the t-distribution depends 
on the degrees of freedom (df), which is determined by the sample size. As the sample size increases, the t-distribution approaches the standard 
normal distribution.

The t-distribution is used in the following situations:

Small Sample Size: When dealing with small sample sizes (typically less than 30), the t-distribution is used instead of the standard normal
distribution. This is because the standard normal distribution assumes knowledge of the population standard deviation, which is often unknown in 
practice.

Unknown Population Standard Deviation: When the population standard deviation is unknown, the t-distribution is used to estimate the population mean.
This commonly occurs when working with limited data or when conducting experiments.

Testing Hypotheses: In hypothesis testing, when comparing means or conducting t-tests, the t-distribution is used to determine the critical region 
and calculate p-values. This allows for the assessment of whether observed differences between groups or variables are statistically significant.

Confidence Interval Estimation: When estimating the confidence interval for the population mean using sample data, the t-distribution is employed. 
The t-distribution provides more conservative intervals compared to the standard normal distribution due to the added uncertainty introduced by 
small sample sizes.
```

```python
Q12: What is t-statistic? State the formula for t-statistic.
Ans.: 
The t-statistic, also known as the t-value, is a test statistic used in hypothesis testing to assess the significance of the difference between 
sample means or to determine the relationship between a sample mean and a known or hypothesized population mean. It quantifies how far the sample 
mean deviates from the null hypothesis mean in terms of standard errors.

The formula for calculating the t-statistic depends on the specific situation and the type of t-test being conducted. Here are the formulas for some
common cases:

One-sample t-test:
The one-sample t-test compares the sample mean to a known or hypothesized population mean.

t = (sample mean - population mean) / (sample standard deviation / sqrt(sample size))

In this formula, the numerator represents the difference between the sample mean and the population mean. The denominator represents the standard 
error of the sample mean, which is the sample standard deviation divided by the square root of the sample size.

Independent two-sample t-test:
The independent two-sample t-test compares the means of two independent samples.

t = (sample mean 1 - sample mean 2) / sqrt((sample variance 1 / sample size 1) + (sample variance 2 / sample size 2))

In this formula, the numerator represents the difference between the means of the two samples. The denominator represents the pooled standard error,
which accounts for the variability within each sample.

Paired two-sample t-test:
The paired two-sample t-test compares the means of two related samples or measures taken on the same subjects.

t = (sample mean difference) / (standard deviation of the differences / sqrt(sample size))

In this formula, the numerator represents the mean difference between the paired observations. The denominator represents the standard error of the
mean difference, which is the standard deviation of the differences divided by the square root of the sample size.
```

```python
Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random sample of 50 days and find the sample mean 
revenue to be $500 with a standard deviation of $50. Estimate the population mean revenue with a 95% confidence interval.
Ans.: To estimate the population mean revenue with a 95% confidence interval, we can use the formula for a confidence interval for a population 
mean when the population standard deviation is unknown. The formula is:

Confidence Interval = sample mean ± (t-value * standard error)

Where:

sample mean: $500 (given)
standard deviation: $50 (given)
sample size: 50 (given)
t-value: corresponds to the desired confidence level and degrees of freedom
standard error: sample standard deviation divided by the square root of the sample size
Since the population standard deviation is unknown, we use the t-distribution instead of the normal distribution. For a 95% confidence level with 
a sample size of 50 (df = 49), the t-value can be obtained from the t-distribution table or using a statistical software.

Assuming the t-value for a 95% confidence level with 49 degrees of freedom is approximately 2.009 (obtained from the t-distribution table), we can
calculate the confidence interval as follows:

Standard Error = standard deviation / sqrt(sample size)
= $50 / sqrt(50)
= $7.07 (approximately)

Confidence Interval = $500 ± (2.009 * $7.07)
= $500 ± $14.21
= ($485.79, $514.21)

Therefore, with a 95% confidence level, we can estimate that the population mean revenue for the coffee shop is between approximately $485.79 and
$514.21.
```

```python
Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.
Ans.:
    To test the hypothesis that the new drug decreases blood pressure by 10 mmHg, we can use a one-sample t-test. Here's how we can conduct the
    hypothesis test with a significance level of 0.05:

Null Hypothesis (H0): The new drug does not decrease blood pressure by 10 mmHg.
Alternative Hypothesis (H1): The new drug decreases blood pressure by 10 mmHg.

Given:

Sample mean decrease in blood pressure: 8 mmHg
Sample standard deviation: 3 mmHg
Sample size: 100
Significance level (alpha): 0.05
Step 1: Set up the hypotheses:
H0: μ = 10 (population mean decrease in blood pressure is 10 mmHg)
H1: μ < 10 (population mean decrease in blood pressure is less than 10 mmHg)

Step 2: Calculate the test statistic:
t = (sample mean - hypothesized mean) / (sample standard deviation / sqrt(sample size))
= (8 - 10) / (3 / sqrt(100))
= -2 / (3/10)
= -20/3
≈ -6.67

Step 3: Determine the critical region:
Since the alternative hypothesis is one-sided (decrease in blood pressure is less than 10 mmHg), we need to find the critical t-value corresponding
to a significance level of 0.05 and degrees of freedom (df) = sample size - 1 = 100 - 1 = 99.

From a t-distribution table or statistical software, the critical t-value for a significance level of 0.05 with 99 degrees of freedom is 
approximately -1.660 (one-sided test).

Step 4: Make a decision:
Since the test statistic (-6.67) is in the critical region (less than -1.660), we reject the null hypothesis.

Step 5: Interpret the result:
With a significance level of 0.05, we have sufficient evidence to conclude that the new drug decreases blood pressure by less than 10 mmHg.
```

```python
Q15. An electronics company produces a certain type of product with a mean weight of 5 pounds and a
standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight
is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5
pounds with a significance level of 0.01.
Ans.:
    To test the hypothesis that the true mean weight of the products is less than 5 pounds, we can use a one-sample t-test. Here's how we can 
    conduct the hypothesis test with a significance level of 0.01:

Null Hypothesis (H0): The true mean weight of the products is equal to 5 pounds.
Alternative Hypothesis (H1): The true mean weight of the products is less than 5 pounds.

Given:

Population mean weight: 5 pounds
Population standard deviation: 0.5 pounds
Sample mean weight: 4.8 pounds
Sample size: 25
Significance level (alpha): 0.01
Step 1: Set up the hypotheses:
H0: μ = 5 (true mean weight is 5 pounds)
H1: μ < 5 (true mean weight is less than 5 pounds)

Step 2: Calculate the test statistic:
t = (sample mean - hypothesized mean) / (sample standard deviation / sqrt(sample size))
= (4.8 - 5) / (0.5 / sqrt(25))
= -0.2 / (0.5 / 5)
= -0.2 / 0.1
= -2

Step 3: Determine the critical region:
Since the alternative hypothesis is one-sided (mean weight is less than 5 pounds), we need to find the critical t-value corresponding to a
significance level of 0.01 and degrees of freedom (df) = sample size - 1 = 25 - 1 = 24.

From a t-distribution table or statistical software, the critical t-value for a significance level of 0.01 with 24 degrees of freedom is 
approximately -2.492 (one-sided test).

Step 4: Make a decision:
Since the test statistic (-2) is in the critical region (less than -2.492), we reject the null hypothesis.

Step 5: Interpret the result:
With a significance level of 0.01, we have sufficient evidence to conclude that the true mean weight of the products is less than 5 pounds.
```

```python
Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 =
30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean
score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two
groups are equal with a significance level of 0.01.
Ans.: To test the hypothesis that the population means for the two groups are equal, we can use a two-sample t-test for independent samples. 
Here's how we can conduct the hypothesis test with a significance level of 0.01:

Null Hypothesis (H0): The population means for the two groups are equal.
Alternative Hypothesis (H1): The population means for the two groups are not equal.

Given:

Sample mean for Group 1: 80
Sample standard deviation for Group 1: 10
Sample size for Group 1: 30
Sample mean for Group 2: 75
Sample standard deviation for Group 2: 8
Sample size for Group 2: 40
Significance level (alpha): 0.01
Step 1: Set up the hypotheses:
H0: μ1 = μ2 (population means for the two groups are equal)
H1: μ1 ≠ μ2 (population means for the two groups are not equal)

Step 2: Calculate the test statistic:
t = (sample mean 1 - sample mean 2) / sqrt((sample variance 1 / sample size 1) + (sample variance 2 / sample size 2))
= (80 - 75) / sqrt((10^2 / 30) + (8^2 / 40))
= 5 / sqrt((100/30) + (64/40))
= 5 / sqrt(3.33 + 1.6)
= 5 / sqrt(4.93)
≈ 5 / 2.22
≈ 2.25

Step 3: Determine the critical region:
Since the alternative hypothesis is two-sided (population means are not equal), we need to find the critical t-value corresponding to a significance
level of 0.01 and degrees of freedom (df) = (sample size 1 - 1) + (sample size 2 - 1) = 30 - 1 + 40 - 1 = 68.

From a t-distribution table or statistical software, the critical t-value for a significance level of 0.01 with 68 degrees of freedom is 
approximately ±2.617 (two-sided test).

Step 4: Make a decision:
Since the test statistic (2.25) does not fall in the critical region (outside ±2.617), we fail to reject the null hypothesis.

Step 5: Interpret the result:
With a significance level of 0.01, we do not have sufficient evidence to conclude that the population means for the two groups are different.
```

```python
Q17. A marketing company wants to estimate the average number of ads watched by viewers during a TV
program. They take a random sample of 50 viewers and find that the sample mean is 4 with a standard
deviation of 1.5. Estimate the population mean with a 99% confidence interval.
Ans.: To estimate the population mean number of ads watched by viewers during a TV program with a 99% confidence interval, we can use the 
formula for a confidence interval for a population mean when the population standard deviation is unknown. The formula is:

Confidence Interval = sample mean ± (t-value * standard error)

Where:

sample mean: 4 (given)
standard deviation: 1.5 (given)
sample size: 50 (given)
t-value: corresponds to the desired confidence level and degrees of freedom
standard error: sample standard deviation divided by the square root of the sample size
Since the population standard deviation is unknown, we use the t-distribution instead of the normal distribution. For a 99% confidence level 
with a sample size of 50 (df = 49), the t-value can be obtained from the t-distribution table or using a statistical software.

Assuming the t-value for a 99% confidence level with 49 degrees of freedom is approximately 2.678 (obtained from the t-distribution table), 
we can calculate the confidence interval as follows:

Standard Error = standard deviation / sqrt(sample size)
= 1.5 / sqrt(50)
≈ 0.2121

Confidence Interval = 4 ± (2.678 * 0.2121)
= 4 ± 0.5689
= (3.4311, 4.5689)

Therefore, with a 99% confidence level, we can estimate that the population mean number of ads watched by viewers during a TV program is 
between approximately 3.4311 and 4.5689.
```
