Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

Ans.: To calculate the Pearson correlation coefficient between two variables, you'll need the following information:

1. Data on the amount of time students spend studying for an exam.
2. Data on their final exam scores.

Once you have this data, you can calculate the Pearson correlation coefficient using the following formula:

\[ r = \frac{n(\sum{xy}) - (\sum{x})(\sum{y})}{\sqrt{[n(\sum{x^2}) - (\sum{x})^2][n(\sum{y^2}) - (\sum{y})^2]}} \]

Where:
- \( r \) is the Pearson correlation coefficient.
- \( n \) is the number of data points (students).
- \( \sum{xy} \) is the sum of the product of the corresponding values of x (study time) and y (exam scores).
- \( \sum{x} \) is the sum of all the values of x.
- \( \sum{y} \) is the sum of all the values of y.
- \( \sum{x^2} \) is the sum of the squares of all the values of x.
- \( \sum{y^2} \) is the sum of the squares of all the values of y.

Interpreting the result:
- If the Pearson correlation coefficient (\( r \)) is close to +1, it indicates a strong positive linear relationship between the two variables. This means that as the amount of time spent studying increases, final exam scores tend to increase as well.
- If \( r \) is close to -1, it indicates a strong negative linear relationship. In this case, as the amount of time spent studying increases, final exam scores tend to decrease.
- If \( r \) is close to 0, it indicates little to no linear relationship between the two variables. Studying time and exam scores do not have a strong linear connection.

Additionally, the sign of \( r \) (+ or -) indicates the direction of the relationship. If it's positive, it suggests a positive correlation, and if it's negative, it suggests a negative correlation.

It's essential to keep in mind that correlation does not imply causation. A high correlation between studying time and exam scores doesn't necessarily mean that studying more causes higher scores. Other factors may be at play, and further analysis or experimentation would be needed to establish causation.

Finally, the strength and significance of the correlation should also be considered. Statistical tests can help determine if the correlation coefficient is significantly different from zero, indicating that the relationship is not due to random chance.

Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
Suppose you have collected data on the amount of sleep individuals get each night and their overall job
satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.

Ans.: Spearman's rank correlation coefficient, often denoted as \( \rho \), is used to measure the strength and direction of the monotonic relationship between two variables. Monotonic relationships are those in which the variables tend to move in the same relative direction (i.e., either both increase or both decrease), but they may not necessarily have a linear relationship.

To calculate Spearman's rank correlation, you'll follow these steps:

1. Rank the values in both variables independently, from lowest to highest. Assign a rank of 1 to the smallest value, 2 to the next smallest, and so on.

2. Calculate the difference between the ranks for each pair of data points. Square these differences.

3. Sum up the squared differences.

4. Use the formula to calculate the Spearman's rank correlation coefficient (\( \rho \)):

\[ \rho = 1 - \frac{6\sum{d^2}}{n(n^2 - 1)} \]

Where:
- \( \rho \) is the Spearman's rank correlation coefficient.
- \( d \) is the difference between the ranks of corresponding data points.
- \( n \) is the number of data points.

Interpreting the result:
- If \( \rho \) is close to +1, it indicates a strong monotonic positive relationship. This means that as the amount of sleep individuals get each night increases, their job satisfaction tends to increase monotonically.
- If \( \rho \) is close to -1, it indicates a strong monotonic negative relationship. In this case, as the amount of sleep increases, job satisfaction tends to decrease monotonically.
- If \( \rho \) is close to 0, it suggests little to no monotonic relationship between the two variables.

The sign of \( \rho \) (+ or -) indicates the direction of the monotonic relationship.

As with the Pearson correlation coefficient, it's essential to remember that correlation does not imply causation. A strong Spearman's rank correlation does not necessarily mean that more sleep causes higher job satisfaction or vice versa. Other factors may be involved, and further analysis would be needed to establish causation.

Additionally, the significance of the correlation coefficient should be considered, typically assessed through hypothesis testing to determine if the observed correlation is statistically significant.

   - Use the Pearson correlation formula I mentioned earlier in response to your first question to calculate 'r'. Here's the formula again for reference:

   Calculate 'r' using the exercise hours as 'x' and BMI as 'y'. This will give you a value that represents the linear relationship between exercise hours and BMI.

2. **Spearman's Rank Correlation (ρ):**
   - As before, you need the raw data on the number of hours of exercise per week and the corresponding BMI for each participant.
   - Rank the values for exercise hours and BMI independently, assigning ranks from 1 to 50 based on their respective magnitudes.
   - Calculate the differences between the ranks for exercise hours and BMI for each participant and square these differences.
   - Use the formula for Spearman's rank correlation to calculate 'ρ':

   \[ \rho = 1 - \frac{6\sum{d^2}}{n(n^2 - 1)} \]

   Where 'd' is the difference between the ranks of exercise hours and BMI.

- If Spearman's rank correlation 'ρ' is close to +1 or -1, it indicates a strong monotonic relationship between exercise hours and BMI. The interpretation is the same as for 'r', but this coefficient is more robust to outliers and non-linear relationships.

Keep in mind that 'r' and 'ρ' may not always give the same result. 'ρ' assesses monotonic relationships, while 'r' assesses linear relationships. Depending on the data, one measure might be more appropriate than the other for describing the relationship between exercise hours and BMI.

Q4. A researcher is interested in examining the relationship between the number of hours individuals
spend watching television per day and their level of physical activity. The researcher collected data on
both variables from a sample of 50 participants. Calculate the Pearson correlation coefficient between
these two variables.

Ans.:To calculate the Pearson correlation coefficient (often denoted as 'r') between the number of hours individuals spend watching television per day and their level of physical activity, you'll need the raw data for both variables from the 50 participants. Once you have the data, you can use the following formula to calculate 'r':

\[ r = \frac{n(\sum{xy}) - (\sum{x})(\sum{y})}{\sqrt{[n(\sum{x^2}) - (\sum{x})^2][n(\sum{y^2}) - (\sum{y})^2]}} \]

Where:
- 'r' is the Pearson correlation coefficient.
- 'n' is the number of data points (in this case, 50 participants).
- 'x' represents the number of hours individuals spend watching television per day.
- 'y' represents the level of physical activity.

Here are the steps to calculate 'r':

1. Calculate the sum of the product of 'x' and 'y' for all data points: \(\sum{xy}\).
2. Calculate the sum of all 'x' values: \(\sum{x}\).
3. Calculate the sum of all 'y' values: \(\sum{y}\).
4. Calculate the sum of the squares of all 'x' values: \(\sum{x^2}\).
5. Calculate the sum of the squares of all 'y' values: \(\sum{y^2}\).

Once you have these values, plug them into the formula to calculate 'r'. The resulting 'r' value will range between -1 and +1, where:

- If 'r' is close to +1, it indicates a strong positive linear relationship. In this context, it would mean that as individuals spend more time watching television per day, their level of physical activity tends to increase.
- If 'r' is close to -1, it indicates a strong negative linear relationship. In this case, as individuals spend more time watching television per day, their level of physical activity tends to decrease.
- If 'r' is close to 0, it suggests little to no linear relationship between the two variables, meaning that the time spent watching television per day is not strongly correlated with the level of physical activity.
