```python
Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.
Ans.: A t-test and a z-test are statistical tests used to make inferences about population parameters based on sample data. The main difference 
between the two lies in the information available about the population.

A t-test is appropriate when the population standard deviation is unknown, and the sample size is relatively small (typically less than 30). It is
used to determine if there is a significant difference between the means of two groups or to compare a sample mean to a known population mean when 
the population standard deviation is unknown. For example:

Scenario for t-test: Suppose you want to compare the effectiveness of two different teaching methods (Method A and Method B) in improving students' 
test scores. You randomly select two groups of students, apply Method A to one group and Method B to the other group, and then measure their test 
scores. To determine if there is a significant difference between the means of the two groups, you can perform a t-test.

On the other hand, a z-test is appropriate when the population standard deviation is known or when the sample size is large (typically greater than 
30). It is used to test hypotheses regarding population means or proportions. A z-test assumes that the population follows a normal distribution. 
For example:

Scenario for z-test: Suppose you are conducting an opinion poll to determine the proportion of people in a city who support a particular political 
candidate. You take a large random sample of individuals, ask them about their preference, and calculate the sample proportion. To test whether the 
proportion of supporters significantly differs from a claimed value (e.g., 50%), you can use a z-test.
```

```python
Q2: Differentiate between one-tailed and two-tailed tests.
Ans.:One-tailed and two-tailed tests are types of hypothesis tests used in statistics to determine the significance of a relationship or difference 
between variables. The difference between the two lies in the directionality of the hypothesis being tested.

One-tailed test: Also known as a directional test, a one-tailed test is used when the hypothesis specifies the direction of the expected relationship 
or difference between variables. The hypothesis is either stated as being greater than or less than a certain value. The one-tailed test evaluates 
whether the data supports the hypothesis in that specified direction.
For example, suppose you are testing the hypothesis that a new drug increases the average lifespan of patients. The one-tailed null hypothesis (H0) 
would be: "The new drug has no effect on average lifespan." The one-tailed alternative hypothesis (H1) would be: "The new drug increases average 
lifespan." In this case, a one-tailed test would assess whether the data provides evidence of a significant increase in average lifespan due to the 
new drug.

Two-tailed test: Also known as a non-directional test, a two-tailed test is used when the hypothesis does not specify a particular direction of the 
relationship or difference. The alternative hypothesis in a two-tailed test states that there is a difference between the variables but does not s
pecify whether it is positive or negative. The two-tailed test evaluates whether the data supports the hypothesis that there is a difference, 
regardless of the direction.
Continuing with the previous example, in a two-tailed test, the null hypothesis (H0) would still be: "The new drug has no effect on average lifespan." 
However, the alternative hypothesis (H1) would be: "The new drug has an effect on average lifespan, either increasing or decreasing it." A two-tailed
test would assess whether the data provides evidence of a significant difference in average lifespan due to the new drug, without specifying whether 
it increases or decreases it.
```

```python
Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.
Ans.: In hypothesis testing, Type 1 and Type 2 errors are two types of mistakes that can occur when making decisions based on statistical tests.

Type 1 error (False Positive): A Type 1 error occurs when the null hypothesis (H0) is incorrectly rejected, indicating a significant effect or 
relationship when, in reality, there is no such effect or relationship. It is the error of concluding that there is a difference or association when 
there is none in the population.
Example scenario for Type 1 error: Let's say a pharmaceutical company conducts a clinical trial to test the effectiveness of a new drug for a specific
medical condition. The null hypothesis (H0) in this case would be that the drug has no effect on the condition. If the company incorrectly rejects the
null hypothesis and claims that the drug is effective, but in reality, it is not, this would be a Type 1 error. Patients might be given a drug that is 
not effective, leading to unnecessary costs and potential side effects.

Type 2 error (False Negative): A Type 2 error occurs when the null hypothesis (H0) is incorrectly accepted, failing to detect a true effect or 
relationship that exists in the population. It is the error of failing to conclude a difference or association when one truly exists.
Example scenario for Type 2 error: Continuing with the pharmaceutical example, let's say the drug being tested is genuinely effective in treating 
the medical condition, but the clinical trial fails to detect this effectiveness. In this case, the null hypothesis (H0) is not rejected, leading 
to a Type 2 error. This could result in the drug not being approved or patients being denied access to a potentially beneficial treatment.
```

```python
Q4:  Explain Bayes's theorem with an example.
Ans.: Bayes's theorem is a fundamental concept in probability theory that allows us to update our beliefs or probabilities based on new evidence. 
It mathematically describes how to revise the probability of an event given prior knowledge and new information. The theorem is named after Thomas 
Bayes, an 18th-century mathematician and clergyman.

Bayes's theorem can be stated as:

P(A|B) = (P(B|A) * P(A)) / P(B)

Where:

P(A|B) represents the conditional probability of event A occurring given that event B has occurred.
P(B|A) represents the conditional probability of event B occurring given that event A has occurred.
P(A) and P(B) represent the probabilities of events A and B occurring independently of each other.
To illustrate Bayes's theorem, let's consider an example:

Suppose there is a rare medical condition that affects 1 in every 10,000 people, and a diagnostic test has been developed to detect this condition. 
The test has a sensitivity of 95%, meaning it correctly identifies positive cases 95% of the time, and a specificity of 98%, meaning it correctly 
identifies negative cases 98% of the time.

Let's say you take the test, and it comes back positive. Now you want to know the probability of actually having the medical condition given the 
positive test result.

In this scenario, we can define the events as follows:

A: Having the medical condition.
B: Testing positive.
We are interested in calculating P(A|B), the probability of having the medical condition given a positive test result.

Using Bayes's theorem, we can calculate this probability by combining the sensitivity and specificity of the test with the prior probability of
having the condition:

P(A|B) = (P(B|A) * P(A)) / P(B)

P(B|A) is the sensitivity of the test, which is 0.95.
P(A) is the prior probability of having the medical condition, which is 1/10,000 or 0.0001.
P(B) is the probability of testing positive, which can be calculated using the sensitivity and specificity:

P(B) = P(B|A) * P(A) + P(B|not A) * P(not A)
= 0.95 * 0.0001 + (1 - 0.98) * (1 - 0.0001)
= 0.000095 + 0.0199
= 0.019995

Now we can substitute these values into Bayes's theorem:

P(A|B) = (0.95 * 0.0001) / 0.019995
≈ 0.00476

Therefore, given a positive test result, the probability of actually having the medical condition is approximately 0.476% or 0.00476.
```

```python
Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
Ans.: A confidence interval is a range of values that provides an estimate of the unknown population parameter with a certain level of confidence. 
It gives us a sense of the precision or uncertainty associated with our sample estimate.

To calculate a confidence interval, the following information is needed:

Sample data: A random sample from the population of interest.
Sample statistic: A summary measure calculated from the sample data (e.g., mean, proportion).
Confidence level: The desired level of confidence, typically expressed as a percentage (e.g., 95%, 99%).
The general formula to calculate a confidence interval is:

Confidence Interval = Sample statistic ± Margin of Error

The margin of error is determined by the standard error of the sample statistic and the critical value corresponding to the chosen confidence level. 
The critical value represents the number of standard deviations needed to encompass the desired confidence level.

Let's consider an example:

Suppose you want to estimate the average height of adult males in a particular city. You take a random sample of 100 adult males and measure their 
heights. The sample mean height is 175 cm, and the sample standard deviation is 5 cm.

You want to calculate a 95% confidence interval for the population mean height.

First, you determine the critical value based on the desired confidence level. For a 95% confidence level, the critical value corresponds to a z-score
of approximately 1.96 (assuming a large enough sample size).

Next, you calculate the standard error, which is the standard deviation of the sample mean:

Standard Error = Sample Standard Deviation / √Sample Size
= 5 / √100
= 0.5

Now, you can calculate the margin of error:

Margin of Error = Critical Value * Standard Error
= 1.96 * 0.5
= 0.98

Finally, you construct the confidence interval:

Confidence Interval = Sample Mean ± Margin of Error
= 175 ± 0.98
= (174.02, 175.98)

The interpretation of this 95% confidence interval is that if you were to repeat the sampling process and construct confidence intervals in the same
way, approximately 95% of those intervals would contain the true population mean height.
```

```python
Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the
event's probability and new evidence. Provide a sample problem and solution.
Ans.: Certainly! Let's consider a sample problem to demonstrate the calculation of probabilities using Bayes' theorem.

Problem:
A factory produces two types of products, Product A and Product B. Historically, it has been observed that 20% of the factory's products are 
defective. The factory has two machines, Machine X and Machine Y, responsible for manufacturing these products. Machine X produces 60% of the 
products, while Machine Y produces the remaining 40%. The defective rates for Machine X and Machine Y are 15% and 25%, respectively. Given that a
randomly selected product is defective, what is the probability that it was produced by Machine X?

Solution:
Let's define the events as follows:
A: Product is defective
X: Product was produced by Machine X

We want to calculate P(X|A), the probability that the product was produced by Machine X given that it is defective.

Using Bayes' theorem, we can calculate this probability using the following formula:

P(X|A) = (P(A|X) * P(X)) / P(A)

P(A|X) represents the probability of a product being defective given that it was produced by Machine X, which is 0.15 (15%).
P(X) represents the probability of a product being produced by Machine X, which is 0.60 (60%).
P(A) represents the probability of a product being defective, which can be calculated using the law of total probability:

P(A) = P(A|X) * P(X) + P(A|not X) * P(not X)
= 0.15 * 0.60 + 0.25 * 0.40
= 0.09 + 0.10
= 0.19

Now we can substitute these values into Bayes' theorem:

P(X|A) = (0.15 * 0.60) / 0.19
= 0.09 / 0.19
≈ 0.474

Therefore, given that a randomly selected product is defective, the probability that it was produced by Machine X is approximately 0.474 or 47.4%.
```

```python
Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population
standard deviation of 5. Interpret the results.
Ans.: To calculate the z-score for a data point, we use the formula:

z = (x - μ) / σ

Where:

x is the data point value
μ is the population mean
σ is the population standard deviation
Let's calculate the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5.

z = (75 - 70) / 5
= 5 / 5
= 1

The z-score is 1.

Interpretation:
The z-score of 1 indicates that the data point is 1 standard deviation above the population mean. In other words, the value of 75 is one standard 
deviation higher than the average value of 70 in the population. The positive z-score suggests that the data point is above the average and further 
away from the mean in terms of standard deviation units.

Z-scores are a way to standardize and compare values within a distribution. They provide a measure of how many standard deviations a particular data 
point is away from the mean. Positive z-scores indicate values above the mean, while negative z-scores indicate values below the mean. A z-score of 0
indicates that the data point is at the mean.
```

```python
Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
significantly effective at a 95% confidence level using a t-test.
Ans.: To conduct a hypothesis test to determine if the weight loss drug is significantly effective at a 95% confidence level using a t-test, we need
to set up the null and alternative hypotheses and calculate the t-statistic and p-value.

Null hypothesis (H0): The weight loss drug is not significantly effective. The average weight loss is not different from zero.
Alternative hypothesis (H1): The weight loss drug is significantly effective. The average weight loss is different from zero.

Given:
Sample size (n) = 50
Sample mean (x̄) = 6 pounds
Sample standard deviation (s) = 2.5 pounds
Confidence level = 95%

Since the sample size is greater than 30, we can use a t-test to assess the hypothesis. The formula for calculating the t-statistic is:

t = (x̄ - μ) / (s / √n)

Where:
x̄ is the sample mean
μ is the hypothesized population mean under the null hypothesis
s is the sample standard deviation
n is the sample size

Let's assume the null hypothesis is μ = 0 (i.e., the average weight loss is zero).

t = (6 - 0) / (2.5 / √50)
= 6 / (2.5 / 7.07)
≈ 6 / 0.354
≈ 16.95

Next, we need to determine the critical t-value based on the degrees of freedom (n - 1) and the desired confidence level. With 49 degrees of freedom 
(50 - 1) and a 95% confidence level, the critical t-value can be obtained from a t-distribution table or a statistical software. For a two-tailed 
test, the critical t-value is approximately ±2.01.

Since our calculated t-value (16.95) is greater than the critical t-value (±2.01), we have sufficient evidence to reject the null hypothesis.

Finally, we need to calculate the p-value associated with the t-statistic to determine the level of significance. The p-value represents the
probability of observing a t-value as extreme or more extreme than the calculated t-value, assuming the null hypothesis is true.

The p-value for a two-tailed test can be obtained from the t-distribution or using statistical software. In this case, the p-value is extremely 
small, likely approaching 0. Therefore, we can conclude that the weight loss drug is significantly effective in producing weight loss based on the 
small p-value and the rejection of the null hypothesis.
```

```python
Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95%
confidence interval for the true proportion of people who are satisfied with their job.
Ans.: To calculate the 95% confidence interval for the true proportion of people who are satisfied with their job, we can use the following formula:

Confidence Interval = Sample Proportion ± (Critical Value * Standard Error)

Where:

Sample Proportion is the proportion of people in the sample who reported being satisfied with their job.
Critical Value is obtained from the standard normal distribution corresponding to the desired confidence level. For a 95% confidence level, the 
critical value is approximately 1.96.
Standard Error is the standard deviation of the sampling distribution, which can be calculated as the square root of 
[(Sample Proportion * (1 - Sample Proportion)) / Sample Size].
Given:
Sample Proportion = 0.65
Sample Size = 500
Critical Value (z-score) for 95% confidence level ≈ 1.96

Let's calculate the confidence interval:

Standard Error = √[(Sample Proportion * (1 - Sample Proportion)) / Sample Size]
= √[(0.65 * 0.35) / 500]
≈ √[0.2275 / 500]
≈ √0.000455
≈ 0.0213

Confidence Interval = Sample Proportion ± (Critical Value * Standard Error)
= 0.65 ± (1.96 * 0.0213)
= 0.65 ± 0.0417
= (0.6083, 0.6917)

The 95% confidence interval for the true proportion of people who are satisfied with their job is approximately (0.6083, 0.6917).
```

```python
Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean
of 65. Calculate the 90% confidence interval for the true population mean.
Ans.: To calculate the 90% confidence interval for the true population mean, we can use the following formula:

Confidence Interval = Sample Mean ± (Critical Value * Standard Error)

Where:

Sample Mean is the mean of the sample observations.
Critical Value is obtained from the t-distribution or Z-table corresponding to the desired confidence level and sample size. Since the sample size 
is large (n > 30), we can use the Z-table. For a 90% confidence level, the critical value is approximately 1.645.
Standard Error is the standard deviation of the sampling distribution, which can be calculated as the population standard deviation divided by the 
square root of the sample size.
Given:
Population Mean (μ) = 60
Population Standard Deviation (σ) = 8
Sample Size (n) = 50
Sample Mean (x̄) = 65
Critical Value (Z) for 90% confidence level ≈ 1.645

Let's calculate the confidence interval:

Standard Error = σ / √n
= 8 / √50
≈ 8 / 7.071
≈ 1.131

Confidence Interval = Sample Mean ± (Critical Value * Standard Error)
= 65 ± (1.645 * 1.131)
= 65 ± 1.86
= (63.14, 66.86)

The 90% confidence interval for the true population mean is approximately (63.14, 66.86).
```
