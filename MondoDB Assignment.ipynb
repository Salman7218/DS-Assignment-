```python
"""Q2. State and Explain the features of MongoDB.
Ans.: MongoDB is a popular open-source NoSQL database that comes with a range of features that make it a popular choice for developers. Here are some 
of the key features of MongoDB:

1. Document-oriented data model: MongoDB uses a document-oriented data model, where data is stored as JSON-like documents with dynamic schemas. This 
makes it easy to store and retrieve complex and unstructured data.

2. Scalability: MongoDB is highly scalable and can be easily scaled horizontally by adding more nodes to a cluster. This makes it ideal for handling 
large-scale applications and databases.

3. High availability: MongoDB supports automatic failover and replica sets, ensuring that the database is always available, even in the event of a node
failure.

4. Indexing and querying: MongoDB provides powerful indexing and querying capabilities, making it easy to search and analyze large volumes of data.

5. Aggregation: MongoDB supports aggregation queries, which enable developers to perform complex data analysis and aggregation operations in real-time.

6. Ad hoc queries: MongoDB supports ad hoc queries, allowing developers to quickly and easily retrieve data without the need to define a predefined 
schema.

7. Flexible schema: MongoDB's flexible schema allows developers to easily add or remove fields from a document without affecting other documents in 
the database.

8. Automatic sharding: MongoDB supports automatic sharding, which allows developers to partition data across multiple nodes in a cluster, improving 
scalability and performance.

9. Geographic distribution: MongoDB supports multi-region and global clusters, making it easy to distribute data across multiple regions for improved 
performance and availability.
```

```python
"""Q4. Using the database and the collection created in question number 3, write a code to insert one record,
and insert many records. Use the find() and find_one() methods to print the inserted record.
Ans.: """

# Import the PyMongo library
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient()
db = client['mydatabase']
collection = db['mycollection']

# Insert one record
document = {'name': 'Alice', 'age': 25, 'email': 'alice@example.com'}
result = collection.insert_one(document)
print('Inserted document with ID:', result.inserted_id)

# Insert many records
documents = [
    {'name': 'Bob', 'age': 35, 'email': 'bob@example.com'},
    {'name': 'Charlie', 'age': 45, 'email': 'charlie@example.com'},
    {'name': 'Dave', 'age': 55, 'email': 'dave@example.com'}
]
result = collection.insert_many(documents)
print('Inserted documents with IDs:', result.inserted_ids)

# Print the inserted record using find_one()
record = collection.find_one({'name': 'Alice'})
print('Found record:', record)

# Print all inserted records using find()
records = collection.find()
for record in records:
    print('Found record:', record)
```

```python
"""Q5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to
demonstrate this.
Ans.: The find() method in MongoDB is used to query the database and retrieve documents that match certain criteria. The find() method 
takes an optional query parameter, which is a dictionary of key-value pairs that represent the search criteria. MongoDB supports a wide 
range of query operators that can be used to specify more complex search criteria.

Here's an example code that demonstrates how to use the find() method to query the "mycollection" collection in the "mydatabase" database 
for documents with an age greater than or equal to 30.
"""
# Import the PyMongo library
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient()
db = client['mydatabase']
collection = db['mycollection']

# Query the database using find()
query = {'age': {'$gte': 30}}
results = collection.find(query)

# Print the results
for result in results:
    print(result)
```

```python
"""Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.
Ans.:The sort() method in MongoDB is used to sort the documents in a collection based on one or more fields. The sort() method takes a 
dictionary of key-value pairs, where the keys are the names of the fields to sort by, and the values are either 1 (to sort in ascending 
order) or -1 (to sort indescending order).

Here's an example code that demonstrates how to use the sort() method to sort the "mycollection" collection in the "mydatabase" database 
by age in descending order:
"""
# Import the PyMongo library
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient()
db = client['mydatabase']
collection = db['mycollection']

# Sort the documents using sort()
results = collection.find().sort('age', -1)

# Print the sorted results
for result in results:
    print(result)

    """In this code, we first connect to the MongoDB database and collection. We then call the find() method on the collection with no 
    arguments to retrieve all documents in the collection, and then call the sort() method on the resulting cursor object to sort the documents
    by age in descending order (-1). This returns a new cursor object with the sorted documents, which we can then iterate over using a loop
    to print each document.
    
    Note that the sort() method can be used to sort by multiple fields by passing in a list of key-value pairs. For example, to sort by age
    in descending order and then by name in ascending order, we would use the following code:"""
    
    results = collection.find().sort([('age', -1), ('name', 1)])

```

```python
"""Q7. Explain why delete_one(), delete_many(), and drop() is used.
Ans.: In MongoDB, the delete_one() and delete_many() methods are used to remove one or multiple documents from a collection, respectively,
based on a specified criteria. The drop() method is used to completely remove a collection from a database.

The delete_one() method removes a single document that matches the specified criteria. If there are multiple documents that match the 
criteria, only the first one encountered will be deleted. Here's an example:
"""
# Delete a single document from the collection
result = collection.delete_one({'name': 'John Doe'})
print(result.deleted_count)  # Prints the number of deleted documents (should be 1)

"""The delete_many() method, on the other hand, removes all documents that match the specified criteria. Here's an example:"""

# Delete all documents from the collection where the age is greater than or equal to 30
result = collection.delete_many({'age': {'$gte': 30}})
print(result.deleted_count)  # Prints the number of deleted documents

"""The drop() method, as mentioned earlier, completely removes a collection from a database. This is useful when you no longer need a 
collection and want to free up the associated storage space. Here's an example:"""

# Drop the collection from the database
collection.drop()

```
